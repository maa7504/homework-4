## Overview
This notebook interacts with the **Google Gemini API** using Python code — not the web interface.

We send **two different prompts** to the model and inspect the full response objects, including:
- The generated text output
- Token usage (prompt tokens, response tokens, total)
- Finish reason
- The full raw candidate object returned by the API

This demonstrates how LLMs can be used programmatically at scale.

## Step 3 — Import and Configure

```python
import google.generativeai as genai

API_KEY = os.environ.get("GEMINI_API_KEY")
genai.configure(api_key=API_KEY)

MODEL = "gemini-1.5-flash"  # free-tier model

print(f"Configured. Using model: {MODEL}")
```

## Step 4 — Define the Two Experiments

We test two different types of prompts:
| # | Type | Goal |
|---|------|------|
| 1 | Creative / Open-ended | See how the model handles a fun, generative task with a persona |
| 2 | Analytical / Factual  | See how the model handles a constrained technical explanation |

```python
EXPERIMENTS = [
    {
        "id": 1,
        "label": "Creative / Open-ended",
        "system": "You are a witty sports journalist who specializes in combat sports.",
        "user":   "Write a short two-sentence hype intro for a Muay Thai championship match.",
    },
    {
        "id": 2,
        "label": "Analytical / Factual",
        "system": "You are a concise computer science tutor.",
        "user":   "Explain the difference between a list and a tuple in Python. Keep it under 80 words.",
    },
]

print(f"Loaded {len(EXPERIMENTS)} experiments.")
```

## Step 5 — Helper: Inspect a Response

```python
def header(title):
    bar = "=" * 60
    print(f"\n{bar}\n  {title}\n{bar}")

def inspect_response(exp, response):
    header(f"Experiment {exp['id']}: {exp['label']}")

    print(f"\n[SYSTEM INSTRUCTION]\n{exp['system']}")
    print(f"\n[USER PROMPT]\n{exp['user']}")
    print(f"\n[OUTPUT TEXT]\n{response.text}")

    print("\n[RESPONSE METADATA]")
    print(f"  Model           : {MODEL}")
    print(f"  Prompt tokens   : {response.usage_metadata.prompt_token_count}")
    print(f"  Response tokens : {response.usage_metadata.candidates_token_count}")
    print(f"  Total tokens    : {response.usage_metadata.total_token_count}")
    print(f"  Finish reason   : {response.candidates[0].finish_reason}")

    print("\n[FULL CANDIDATE OBJECT]")
    print(response.candidates[0])

print("Helper functions defined.")
```

## Step 6 — Run Both Experiments and Inspect Outputs

```python
print("LLM API Demo — Google Gemini")
print(f"Using model: {MODEL}\n")

for exp in EXPERIMENTS:
    # Initialize model with a system instruction (persona)
    model = genai.GenerativeModel(
        model_name=MODEL,
        system_instruction=exp["system"],
    )

    # Send the user prompt and get a response
    response = model.generate_content(exp["user"])

    # Inspect and print the full output
    inspect_response(exp, response)

header("Done — both experiments complete.")
```
